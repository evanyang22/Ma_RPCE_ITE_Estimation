```python
#Looping through all the replications of the IHDP TensorDataset

#these store the averages over 60 test samples
unconfounded=[]
confounded= []
noiseless=[]
noised = []
confidence=[]
RPCE_CATE=[]
treatment=[]
yf=[]
ycf= []
mu1=[]
mu0=[]
for i in range(0,100):
    print ("On Replication " + str(i))
    #Creating training and testing Tensor datasets of observational and RCT data
    ihdp_train_rct, ihdp_train_obs = create_ihdp_rct_obs_datasets(
        "C:\\Users\\evany\\Desktop\\MaLabRotation\\Data\\FredJoData\\ihdp_npci_1-100.train.npz",
        replication=i
    )

    ihdp_test_rct, ihdp_test_obs = create_ihdp_rct_obs_datasets(
        "C:\\Users\\evany\\Desktop\\MaLabRotation\\Data\\FredJoData\\ihdp_npci_1-100.test.npz",
        replication=i, rct_fraction=0.0
    )
    # New
    # ---- NORMALIZE OUTCOMES PER REPLICATION ----
    # Pool observed outcomes from both splits to get stable statistics
    y_obs = ihdp_train_obs.tensors[2]   # yf from OBS
    y_rct = ihdp_train_rct.tensors[2]   # yf from RCT
    y_all = torch.cat([y_obs, y_rct])
    y_mean = y_all.mean()
    y_std  = y_all.std().clamp_min(1e-8)

    def normalize_dataset(ds, y_mean, y_std):
        """Normalize yf, ycf, mu0, mu1 — leave X and T untouched."""
        tensors = list(ds.tensors)
        tensors[2] = (tensors[2] - y_mean) / y_std   # yf
        tensors[3] = (tensors[3] - y_mean) / y_std   # ycf
        tensors[4] = (tensors[4] - y_mean) / y_std   # mu0
        tensors[5] = (tensors[5] - y_mean) / y_std   # mu1
        return TensorDataset(*tensors)
    ihdp_train_obs_n = normalize_dataset(ihdp_train_obs, y_mean, y_std)
    ihdp_train_rct_n = normalize_dataset(ihdp_train_rct, y_mean, y_std)
    ihdp_test_obs_n  = normalize_dataset(ihdp_test_obs,  y_mean, y_std)
    #End New
    
    AutoEncoderModel, EncoderTrainHistory, binaryIHDP, continuousIHDP= train_mixed_autoencoder(TensorDataset(ihdp_train_obs_n.tensors[0],ihdp_train_obs_n.tensors[1],ihdp_train_obs_n.tensors[2])
                                                                                               ,hidden_dim=64,latent_dim=16, num_epochs=100, verbose=False, outcome_type= "continuous")
    initialize_stage2_from_stage1(AutoEncoderModel)
    
    AutoEncoderModel, stage2_history = train_stage2_rct(
        model=AutoEncoderModel,
        dataset=TensorDataset(ihdp_train_rct_n.tensors[0],ihdp_train_rct_n.tensors[1],ihdp_train_rct_n.tensors[2]),
        batch_size=32,
        lr=5e-4,
        num_epochs=150,
        lambda_shift=1e-3,
        outcome_type="continuous",
        freeze_encoder=True,
        encoder_lr_ratio=0.05,
        lambda_encoder_shift=1e-3,
        verbose=False
    )

    #was ihdp_test_rct.tensors[0], not ihdp_train_rct.tensors[0]
    ihdp_cate, ihdp_cost,unconfoundedCATE, confoundedCATE =predict_cate_rpce(AutoEncoderModel, ihdp_test_obs_n.tensors[0], ihdp_train_rct_n.tensors[0],outcome_type="continuous")


    #New
     # ---- RESCALE CATE BACK TO ORIGINAL UNITS ----
    # CATE = Y1 - Y0.  Since Y_norm = (Y - mean) / std,
    # CATE_norm = (Y1 - mean)/std - (Y0 - mean)/std = (Y1 - Y0)/std
    # So: CATE_original = CATE_norm * std
    ihdp_cate = ihdp_cate * y_std
    unconfoundedCATE = unconfoundedCATE * y_std
    confoundedCATE = confoundedCATE * y_std

    #End New
    treatment.append(ihdp_test_obs.tensors[1])
    
    yf.append(ihdp_test_obs.tensors[2])
    
    ycf.append(ihdp_test_obs.tensors[3])

    mu1.append(ihdp_test_obs.tensors[5])
    mu0.append(ihdp_test_obs.tensors[4])
    
    
    print(" Noiseless Potential Outcome = " +str(torch.mean(ihdp_test_obs.tensors[5]- ihdp_test_obs.tensors[4])))
    noiseless.append(ihdp_test_obs.tensors[5]- ihdp_test_obs.tensors[4])
   
    
    #print(" Noised Potential Outcome = " +str(torch.mean(abs(ihdp_test_obs.tensors[3]- ihdp_test_obs.tensors[2]))))
    #noised.append(abs(ihdp_test_obs.tensors[3]- ihdp_test_obs.tensors[2]))
    
    noisedSection=[] 
    for k in range(0,75):
        if ihdp_test_obs.tensors[1][k] ==1:
            noisedSection.append(ihdp_test_obs.tensors[2][k]- ihdp_test_obs.tensors[3][k])
        else:
            noisedSection.append(ihdp_test_obs.tensors[3][k]- ihdp_test_obs.tensors[2][k])

    noisedSection=torch.tensor(noisedSection)
    print(" Noised Potential Outcome = " +str(torch.mean(noisedSection)))
    noised.append(noisedSection)
    print(" Average Confidence = " +str(torch.mean(ihdp_cost)))
    confidence.append(ihdp_cost)
    #print(" Average Unconfounded CATE = " + str(unconfoundedCATE))
    print(" Average Unconfounded CATE = " + str(torch.mean(unconfoundedCATE)))
    unconfounded.append(unconfoundedCATE)
    #print(len(unconfoundedCATE))
    #unconfounded.append(
    print(" Average Confounded CATE = " + str(torch.mean(confoundedCATE)))
    confounded.append(confoundedCATE)
    print(" Average Predicted (RPCE-Model) CATE = " +str(torch.mean( ihdp_cate)))
    RPCE_CATE.append(ihdp_cate)
```

```python
"""
DIAGNOSTIC: Encoder-shift vs Data-scarcity
 
Run this as a notebook cell. It tests 3 conditions on the SAME replications
so results are directly comparable:
 
  A) frozen encoder    — isolates Stage 2 head quality with stable z
  B) unfrozen encoder  — your current setup
  C) oracle ceiling    — trains g0/g1 on FULL obs+rct data (not realistic,
                          but shows what's achievable if data weren't scarce)
 
If A >> B:  encoder shift is the problem (unfreezing hurts)
If A ≈ B and both bad:  data scarcity is the problem (75 samples isn't enough)
If C >> A:  data scarcity is confirmed (more data = much better)
If C ≈ A:  the representation itself is the bottleneck
"""
 
import torch
import numpy as np
from torch.utils.data import TensorDataset
 
# ---- CONFIG ----
# Adjust these paths to match your setup
TRAIN_PATH = "C:\\Users\\evany\\Desktop\\MaLabRotation\\Data\\FredJoData\\ihdp_npci_1-100.train.npz"
TEST_PATH  = "C:\\Users\\evany\\Desktop\\MaLabRotation\\Data\\FredJoData\\ihdp_npci_1-100.test.npz"
 
# Test on a subset of replications to save time
TEST_REPS = [0, 8, 20, 27, 33, 44, 50, 67, 80, 97]  # mix of good and bad
 
results_frozen = []
results_unfrozen = []
results_oracle = []
 
for i in TEST_REPS:
    print(f"\n{'='*60}")
    print(f"Replication {i}")
    print(f"{'='*60}")
 
    # ---- Create datasets ----
    ihdp_train_rct, ihdp_train_obs = create_ihdp_rct_obs_datasets(TRAIN_PATH, replication=i)
    ihdp_test_rct, ihdp_test_obs = create_ihdp_rct_obs_datasets(TEST_PATH, replication=i, rct_fraction=0.0)
 
    # ---- Normalize ----
    y_obs = ihdp_train_obs.tensors[2]
    y_rct = ihdp_train_rct.tensors[2]
    y_all = torch.cat([y_obs, y_rct])
    y_mean = y_all.mean()
    y_std = y_all.std().clamp_min(1e-8)
 
    def normalize_dataset(ds, y_mean, y_std):
        tensors = list(ds.tensors)
        tensors[2] = (tensors[2] - y_mean) / y_std
        tensors[3] = (tensors[3] - y_mean) / y_std
        tensors[4] = (tensors[4] - y_mean) / y_std
        tensors[5] = (tensors[5] - y_mean) / y_std
        return TensorDataset(*tensors)
 
    train_obs_n = normalize_dataset(ihdp_train_obs, y_mean, y_std)
    train_rct_n = normalize_dataset(ihdp_train_rct, y_mean, y_std)
    test_obs_n = normalize_dataset(ihdp_test_obs, y_mean, y_std)
 
    # Ground truth for evaluation
    true_ite = ihdp_test_obs.tensors[5] - ihdp_test_obs.tensors[4]  # mu1 - mu0, original scale
    true_ate = true_ite.mean().item()
 
    def evaluate_branches(model, test_obs_n, train_rct_n, y_std, true_ite, true_ate, label):
        """Run CATE prediction and report unconfounded vs confounded quality."""
        cate_rpce, conf, cate_unconf, cate_conf = predict_cate_rpce(
            model, test_obs_n.tensors[0], train_rct_n.tensors[0],
            outcome_type="continuous"
        )
        # Rescale to original units
        cate_rpce = cate_rpce * y_std
        cate_unconf = cate_unconf * y_std
        cate_conf = cate_conf * y_std
 
        ate_err_unconf = abs(cate_unconf.mean().item() - true_ate)
        ate_err_conf = abs(cate_conf.mean().item() - true_ate)
        pehe_unconf = torch.sqrt(((cate_unconf - true_ite)**2).mean()).item()
        pehe_conf = torch.sqrt(((cate_conf - true_ite)**2).mean()).item()
        mean_conf = conf.mean().item()
 
        print(f"  [{label}]")
        print(f"    Unconf ATE err: {ate_err_unconf:.4f}  PEHE: {pehe_unconf:.4f}")
        print(f"    Conf   ATE err: {ate_err_conf:.4f}  PEHE: {pehe_conf:.4f}")
        print(f"    Mean confidence: {mean_conf:.4f}")
 
        return {
            'ate_unconf': ate_err_unconf, 'ate_conf': ate_err_conf,
            'pehe_unconf': pehe_unconf, 'pehe_conf': pehe_conf,
            'mean_conf': mean_conf
        }
 
    # ============================================================
    # CONDITION A: FROZEN ENCODER (isolates data scarcity)
    # ============================================================
    model_A, _, _, _ = train_mixed_autoencoder(
        TensorDataset(train_obs_n.tensors[0], train_obs_n.tensors[1], train_obs_n.tensors[2]),
        hidden_dim=64, latent_dim=16, num_epochs=100, lr=1e-3,
        verbose=False, outcome_type="continuous"
    )
    initialize_stage2_from_stage1(model_A)
    model_A, _ = train_stage2_rct(
        model=model_A,
        dataset=TensorDataset(train_rct_n.tensors[0], train_rct_n.tensors[1], train_rct_n.tensors[2]),
        batch_size=32, lr=5e-4, num_epochs=150, lambda_shift=1e-3,
        outcome_type="continuous",
        freeze_encoder=True,  # <-- FROZEN
        verbose=False
    )
    res_A = evaluate_branches(model_A, test_obs_n, train_rct_n, y_std, true_ite, true_ate, "FROZEN")
    results_frozen.append(res_A)
 
    # ============================================================
    # CONDITION B: UNFROZEN ENCODER (your current setup)
    # ============================================================
    model_B, _, _, _ = train_mixed_autoencoder(
        TensorDataset(train_obs_n.tensors[0], train_obs_n.tensors[1], train_obs_n.tensors[2]),
        hidden_dim=64, latent_dim=16, num_epochs=100, lr=1e-3,
        verbose=False, outcome_type="continuous"
    )
    initialize_stage2_from_stage1(model_B)
    model_B, _ = train_stage2_rct(
        model=model_B,
        dataset=TensorDataset(train_rct_n.tensors[0], train_rct_n.tensors[1], train_rct_n.tensors[2]),
        batch_size=32, lr=5e-4, num_epochs=150, lambda_shift=1e-3,
        outcome_type="continuous",
        freeze_encoder=False,  # <-- UNFROZEN
        encoder_lr_ratio=0.05,
        lambda_encoder_shift=1e-3,
        verbose=False
    )
    res_B = evaluate_branches(model_B, test_obs_n, train_rct_n, y_std, true_ite, true_ate, "UNFROZEN")
    results_unfrozen.append(res_B)
 
    # ============================================================
    # CONDITION C: ORACLE — train g0/g1 on ALL data (OBS + RCT)
    # This is NOT a real method — it cheats by using OBS outcomes
    # for the "unconfounded" heads. It establishes a CEILING for
    # what g0/g1 could achieve if data weren't scarce.
    # ============================================================
    model_C, _, _, _ = train_mixed_autoencoder(
        TensorDataset(train_obs_n.tensors[0], train_obs_n.tensors[1], train_obs_n.tensors[2]),
        hidden_dim=64, latent_dim=16, num_epochs=100, lr=1e-3,
        verbose=False, outcome_type="continuous"
    )
    initialize_stage2_from_stage1(model_C)
    # Combine OBS + RCT for oracle training
    X_all = torch.cat([train_obs_n.tensors[0], train_rct_n.tensors[0]], dim=0)
    T_all = torch.cat([train_obs_n.tensors[1], train_rct_n.tensors[1]], dim=0)
    Y_all = torch.cat([train_obs_n.tensors[2], train_rct_n.tensors[2]], dim=0)
    oracle_ds = TensorDataset(X_all, T_all, Y_all)
    model_C, _ = train_stage2_rct(
        model=model_C, dataset=oracle_ds,
        batch_size=64, lr=5e-4, num_epochs=150, lambda_shift=1e-3,
        outcome_type="continuous",
        freeze_encoder=True,
        verbose=False
    )
    res_C = evaluate_branches(model_C, test_obs_n, train_rct_n, y_std, true_ite, true_ate, "ORACLE")
    results_oracle.append(res_C)
 
 
# ============================================================
# SUMMARY
# ============================================================
print("\n\n" + "=" * 70)
print("SUMMARY ACROSS ALL TEST REPLICATIONS")
print("=" * 70)
 
def summarize(results, label):
    ate = np.mean([r['ate_unconf'] for r in results])
    pehe = np.mean([r['pehe_unconf'] for r in results])
    ate_c = np.mean([r['ate_conf'] for r in results])
    pehe_c = np.mean([r['pehe_conf'] for r in results])
    conf = np.mean([r['mean_conf'] for r in results])
    print(f"\n  [{label}]")
    print(f"    Unconfounded  — ATE err: {ate:.4f}, PEHE: {pehe:.4f}")
    print(f"    Confounded    — ATE err: {ate_c:.4f}, PEHE: {pehe_c:.4f}")
    print(f"    Mean confidence: {conf:.4f}")
    return ate, pehe
 
ate_A, pehe_A = summarize(results_frozen,   "A: FROZEN encoder")
ate_B, pehe_B = summarize(results_unfrozen, "B: UNFROZEN encoder")
ate_C, pehe_C = summarize(results_oracle,   "C: ORACLE (all data)")
 
print("\n" + "-" * 70)
print("DIAGNOSIS:")
if ate_A < ate_B * 0.8:
    print("  → A << B: ENCODER SHIFT is hurting. Keep encoder frozen.")
elif ate_A > ate_B * 1.2:
    print("  → B << A: Unfreezing helps. Encoder shift is not the issue.")
else:
    print("  → A ≈ B: Encoder shift is NOT the main problem.")
 
if ate_C < ate_A * 0.5:
    print("  → C << A: DATA SCARCITY is the main bottleneck.")
    print("    g0/g1 can learn well with more data — 149 RCT samples")
    print("    is simply not enough for this architecture.")
elif ate_C < ate_A * 0.8:
    print("  → C < A: Data scarcity contributes, but representation quality")
    print("    is also a factor.")
else:
    print("  → C ≈ A: More data doesn't help much. The REPRESENTATION")
    print("    is the bottleneck — the encoder doesn't capture what")
    print("    g0/g1 need to predict treatment effects.")
 
print()
print("NOTE: The oracle uses OBS data (confounded) for g0/g1 training,")
print("so it's not unbiased — but it tests whether MORE data helps")
print("the heads learn the response surface at all.")
```
